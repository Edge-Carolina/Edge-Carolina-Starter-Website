# **Getting Started**

## Setting Up Your Git Configuration

As part of the onboarding process, it's important to configure your Git settings to ensure that your commits are properly attributed to you. Follow these steps to set up your Git configuration:

### 1. Configure Your Email and Name

From your root directory navigate to the .devcontainer and select Dockerfile. In this file you will see two different commands:

```bash
git config --global user.email "you@example.com"
```
```bash
git config --global user.name "Your Name"
```
Please replace `you@example.com` and `Your Name` with your email and name respectively.
